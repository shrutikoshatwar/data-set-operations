```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/match_events.csv")
```

```python
print("First 10 Rows")
print(df.head(10))
```

```python
print("\nLast 5 Rows")
print(df.tail())
```

```python
print("\nColumns:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("Shape:", df.shape)
```

```python
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python

print("\nUnique Values")
print(df.nunique())
```

```python
print("Missing Values")
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.describe())

print("\nNumerical Columns")
print(df.select_dtypes(include="number").columns)
```

```python
print(df.select_dtypes(include="object").columns)
```

```python
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="minute", y="player_id", data=df)
plt.title("Minute vs Player ID")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="event_type", y="minute", data=df)
plt.xticks(rotation=45)
plt.title("Minute by Event Type")
plt.show()
```

```python
df["event_type"].value_counts().plot.pie(
    autopct="%1.1f%%",
    figsize=(6,6)
)
plt.ylabel("")
plt.title("Event Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
