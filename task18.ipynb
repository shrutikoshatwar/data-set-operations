```python
import numpy as np
import pandas as pd

from google.colab import files

uploaded = files.upload()
```

```python
filepath = "/content/ford_car_dataset.csv"
df = pd.read_csv(filepath)
```

```python

# Display first 10 rows
print("First 10 Rows:")
print(df.head(10))

# Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail())

# Shape of the dataset
print("\nShape of Dataset:")
print(df.shape)

# Data types of all columns
print("\nData Types:")
print(df.dtypes)

# Dataset information
print("\nDataset Info:")
print(df.info())
```

```python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Check duplicate rows
print("\nDuplicate Rows:", df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Shape after removing duplicates
print("\nShape After Removing Duplicates:")
print(df.shape)

# Observation:
# 1. Checked all columns for missing values using isnull().sum().
# 2. Checked duplicate records using duplicated().sum().
# 3. Removed duplicate rows using drop_duplicates().
# 4. The cleaned dataset is now ready for further analysis and preprocessing.
```

```python
# Statistical summary of numerical columns
print("Statistical Summary:")
print(df.describe())

# Price Statistics
print("\nPrice Statistics")
print("Minimum Price :", df["price"].min())
print("Maximum Price :", df["price"].max())
print("Mean Price    :", df["price"].mean())
print("Median Price  :", df["price"].median())

# Mileage Statistics
print("\nMileage Statistics")
print("Minimum Mileage :", df["mileage"].min())
print("Maximum Mileage :", df["mileage"].max())
print("Mean Mileage    :", df["mileage"].mean())
print("Median Mileage  :", df["mileage"].median())

# Year Statistics
print("\nYear Statistics")
print("Minimum Year :", df["year"].min())
print("Maximum Year :", df["year"].max())
print("Mean Year    :", df["year"].mean())
print("Median Year  :", df["year"].median())

# Observation:
# 1. describe() provides count, mean, standard deviation,
#    minimum, maximum and quartile values.
# 2. The average price and mileage help understand the
#    overall characteristics of the cars.
# 3. The minimum and maximum year show the oldest and
#    newest cars available in the dataset.
# 4. The median values represent the central tendency of
#    the numerical features.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
sns.countplot(y="fuelType", data=df, palette="Set2")
plt.title("Fuel Type")

plt.subplot(2,2,2)
sns.countplot(y="transmission", data=df, palette="Set3")
plt.title("Transmission")

plt.subplot(2,2,3)
sns.countplot(y="model", data=df, palette="tab20")
plt.title("Model")

plt.tight_layout()
plt.show()

# Observation:
# 1. The count plots show the frequency of each fuel type,
#    transmission type, and car model.
# 2. The categories with the tallest bars are the most common.
# 3. These plots help identify popular Ford models,
#    commonly used fuel types, and preferred transmissions.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_cols = ["model", "transmission", "fuelType"]

plt.figure(figsize=(15,10))

for i, col in enumerate(cat_cols):
    plt.subplot(2,2,i+1)
    sns.countplot(data=df, x=col, palette="Set2")
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Observation:
# 1. The heatmap shows the correlation between all numerical features.
# 2. Correlation values range from -1 to +1.
# 3. Values close to +1 indicate a strong positive correlation.
# 4. Values close to -1 indicate a strong negative correlation.
# 5. The feature having the highest correlation with 'price'
#    is the most useful for predicting car prices.
```

```python
# Independent Features (Input Features)
X = df.drop("price", axis=1)
print("Independent Features:")
print(X.columns)

# Dependent Feature (Target Feature)
y = df["price"]
print("\nDependent Feature:")
print(y.name)

# Observation:
# 1. Independent Features (X):
#    model, year, transmission, mileage,
#    fuelType, tax, mpg, engineSize
#
# 2. Dependent Feature (y):
#    price
#
# Reason:
# Price is the target variable that we want to predict
# using the remaining features.
```

```python
# Display categorical columns before encoding
print("Before Encoding:")
print(df[["model", "transmission"]].head())

# Apply One Hot Encoding
df_encoded = pd.get_dummies(df, columns=["model", "transmission", "fuelType"], drop_first=True)

# Display encoded columns
print("\nAfter Encoding:")
print(df_encoded.head())

# Observation:
# 1. Categorical columns (model, transmission, fuelType)
#    were converted into numerical format using One Hot Encoding.
# 2. Each category is represented by a separate binary (0/1) column.
# 3. drop_first=True is used to avoid the dummy variable trap.
# 4. The encoded dataset is now suitable for machine learning algorithms.
```

```python
from sklearn.preprocessing import StandardScaler

# Independent Features
X = df.drop("price", axis=1)

# Select only numeric columns
X_num = X.select_dtypes(include=["int64", "float64"])

# Apply Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

# Convert scaled data into DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X_num.columns)

# Display first 5 rows
print("Scaled Data (First 5 Rows):")
print(X_scaled.head())

# Observation:
# 1. StandardScaler standardizes all numeric features.
# 2. The scaled features have a mean close to 0 and a standard deviation close to 1.
# 3. Feature scaling improves the performance of many machine learning algorithms.
# 4. Only numeric independent features were scaled.
```
